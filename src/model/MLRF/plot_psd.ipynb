```python
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
def plot_psd(directory):
    plt.figure(figsize=(15, 20))
    file_counter = 0
    for i, filename in enumerate(os.listdir(directory), 1):
        if filename.endswith('.npy') and 'wifi' in filename.lower() and file_counter < 10:
            data = np.load(os.path.join(directory, filename))
            plt.subplot(5, 2, file_counter + 1)
            plt.plot(data, label=filename)
            plt.legend()
            plt.title(f'Plot of {filename}')
            plt.xlabel('Samples')
            plt.ylabel('dBm')
            file_counter += 1
    plt.tight_layout()
    plt.show()
```

```python
plot_psd('./data/open_data/training_psd/')
plot_psd('./data/recorded_data/training_psd/')
```
